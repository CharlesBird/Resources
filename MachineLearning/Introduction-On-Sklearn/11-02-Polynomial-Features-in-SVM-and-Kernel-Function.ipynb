### 使用多项式特征的SVM

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline

def PolynomialSVC(degree, C=1.0):
    return Pipeline([
        ('Poly', PolynomialFeatures(degree=degree)),
        ('std_scaler', StandardScaler()),
        ('svc', LinearSVC(C=C))
    ])
```

```python
poly_svc = PolynomialSVC(degree=3)
poly_svc.fit(X, y)
```

```python
def plot_decision_boundry(model, axis):
    x0, x1 = np.meshgrid(
        np.linspace(axis[0], axis[1], int((axis[1]-axis[0])*100)).reshape(-1, 1),
        np.linspace(axis[2], axis[3], int((axis[3]-axis[2])*100)).reshape(-1, 1),
        )
    X_new = np.c_[x0.ravel(), x1.ravel()]
    y_predict = model.predict(X_new)
    zz = y_predict.reshape(x0.shape)
    
    from matplotlib.colors import ListedColormap
    custom_cmap = ListedColormap(['#EF9A9A','#FFF59D','#90CAF9'])
    plt.contourf(x0, x1, zz, cmap=custom_cmap)
```

```python
plot_decision_boundry(poly_svc, [-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0, 0], X[y==0, 1], color='red')
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue')
plt.show()
```

```python
poly_svc2 = PolynomialSVC(degree=2)
poly_svc2.fit(X, y)
```

```python
plot_decision_boundry(poly_svc2, [-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0, 0], X[y==0, 1], color='red')
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue')
plt.show()
```

```python
poly_svc3 = PolynomialSVC(degree=4)
poly_svc3.fit(X, y)
```

```python
plot_decision_boundry(poly_svc3, [-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0, 0], X[y==0, 1], color='red')
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue')
plt.show()
```

```python
poly_svc4 = PolynomialSVC(degree=5)
poly_svc4.fit(X, y)
```

```python
plot_decision_boundry(poly_svc4, [-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0, 0], X[y==0, 1], color='red')
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue')
plt.show()
```

```python
poly_svc5 = PolynomialSVC(degree=5, C=0.01)
poly_svc5.fit(X, y)
```

```python
plot_decision_boundry(poly_svc5, [-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0, 0], X[y==0, 1], color='red')
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue')
plt.show()
```

### 使用多项式核函数的SVN

```python
from sklearn.svm import SVC

def PolynomialKernelSVC(degree, C=1.0):
    return Pipeline([
        ('std_scaler', StandardScaler()),
        ('KernelSVC', SVC(kernel="poly", degree=degree, C=C))
    ])
```

```python
poly_kernel_svc = PolynomialKernelSVC(3)
poly_kernel_svc.fit(X, y)
```

```python
plot_decision_boundry(poly_kernel_svc, [-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0, 0], X[y==0, 1], color='red')
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue')
plt.show()
```

```python
poly_kernel_svc2 = PolynomialKernelSVC(5)
poly_kernel_svc2.fit(X, y)
```

```python
plot_decision_boundry(poly_kernel_svc2, [-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0, 0], X[y==0, 1], color='red')
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue')
plt.show()
```

```python
poly_kernel_svc3 = PolynomialKernelSVC(degree=3, C=0.01)
poly_kernel_svc3.fit(X, y)
```

```python
plot_decision_boundry(poly_kernel_svc3, [-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0, 0], X[y==0, 1], color='red')
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue')
plt.show()
```
