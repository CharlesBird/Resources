```python
x = np.linspace(0, 10, 100)
```

```python
siny = np.sin(x)
```

```python
cosy = np.cos(x)
```

```python
plt.plot(x, siny)
```

```python
plt.plot(x, cosy, color='red')
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red')
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='--')
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='--')
plt.xlim(-1, 15)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='--')
plt.xlim(-1, 15)
plt.ylim(-2, 2)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='--')
plt.axis([-1, 15, -2, 2])
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color='red', linestyle='--')
plt.axis([-1, 15, -2, 2])
plt.xlabel('x name')
plt.ylabel('x value')
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color='red', linestyle='--', label="cos(x)")
plt.axis([-1, 15, -2, 2])
plt.xlabel('x name')
plt.ylabel('x value')
plt.legend()
plt.title('Image Name')
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y, alpha=0.5)
plt.show()
```
